```python
def batch_iterator(iterator, batch_size):
    """Returns lists of length batch_size.

    This can be used on any iterator, for example to batch up
    SeqRecord objects from Bio.SeqIO.parse(...), or to batch
    Alignment objects from Bio.AlignIO.parse(...), or simply
    lines from a file handle.

    This is a generator function, and it returns lists of the
    entries from the supplied iterator.  Each list will have
    batch_size entries, although the final list may be shorter.
    """
    entry = True  # Make sure we loop once
    while entry:
        batch = []
        while len(batch) < batch_size:
            try:
                # entry = iterator.next()
                entry = next(iterator)
            except StopIteration:
                entry = None
            if entry is None:
                # End of file
                break
            batch.append(entry)
        if batch:
            yield batch
```

```python
from Bio import SeqIO

record_iter = SeqIO.parse(open("test_disprot_fasta/disprot_homo_sapiens.fasta"), "fasta")
for i, batch in enumerate(batch_iterator(record_iter, 100)):
    filename = "test_disprot_fasta/group_%i.fasta" % (i + 1)
    with open(filename, "w") as handle:
        count = SeqIO.write(batch, handle, "fasta")
    print("Wrote %i records to %s" % (count, filename))
```

# AA relative count plotting "Median with confidence interval"

```python
target_type         = "median"
confidence_interval = (0.1, 0.9)
show                = True
xlim                = (0, 25)
bins                = 100
sharex              = True
sharey              = False

_, medians_gpcr,    left_confs_gpcr,    right_confs_gpcr    = gpcr.grid_hist_proteins_residue_relative_count( df = gpcr.GPCR_DF,
                                                                                             target_type         = target_type,
                                                                                             confidence_interval = confidence_interval,
                                                                                             bins                = bins,
                                                                                             xlim                = xlim,
                                                                                             show                = show,
                                                                                             sharex              = sharex,
                                                                                             sharey              = sharey)
_, medians_disprot, left_confs_disprot, right_confs_disprot = gpcr.grid_hist_proteins_residue_relative_count( df = gpcr.DISPROT_DF,
                                                                                             target_type         = target_type,
                                                                                             confidence_interval = confidence_interval,
                                                                                             bins                = bins,
                                                                                             xlim                = xlim,
                                                                                             show                = show,
                                                                                             sharex              = sharex,
                                                                                             sharey              = sharey )
```

```python
x = medians_gpcr;    x_left   = left_confs_gpcr;    x_right  = right_confs_gpcr
y = medians_disprot; y_bottom = left_confs_disprot; y_top    = right_confs_disprot

xerr_left   = x       - x_left
xerr_right  = x_right - x
yerr_bottom = y       - y_bottom
yerr_top    = y_top   - y

fig = gpcr.line_compare_points( x = x,
                                y = y,
                         xerr                = (xerr_left,   xerr_right),
                         yerr                = (yerr_bottom, yerr_top),
                         figsize             = (15,10),
                         scale_factor        = 0.7,
                         errorbar_capsize    = 0,
                         errorbar_capthick   = 0,
                         scatter_marker_size = 3,
                         legend_location     = "center right",
                         xlim                = (-1, 19),
                         ylim                = (-1, 17),
                         label_xaxis         = "GPCR AA relative count",
                         label_yaxis         = "DisProt AA relative count",
            title_add           = f" (Medians, Confidence Interval: {(confidence_interval[1] - confidence_interval[0])*100:.1f}%)"
                       )
```

```python
def get_fit_line(x, y):
    x     = np.array(x)
    y     = np.array(y)
    m, c  = np.polyfit(x,y,deg=1)
    y_line  = m * x + c
    return y_line
```
